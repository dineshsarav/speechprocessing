# A1

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

y, sr = librosa.load('AI_IN_speechprocessing.wav')
librosa.display.waveshow(y)
plt.title('Speech Signal')
```

# A2

```python
duration = len(y) / sr  # Duration in seconds
print(f"Duration of the audio: {duration:.2f} seconds")
print(f"Magnitude range: {y.min()} to {y.max()}")
```

# A3

```python
import IPython.display as ipd

start_time = 2
end_time = 3

# Extract the segment
segment = y[int(start_time * sr):int(end_time * sr)]

# Display the waveform
librosa.display.waveshow(segment)
plt.title('Speech Signal Segment')

# Play the segment
ipd.Audio(segment, rate=sr)
```

# A4

```python
s1 = 0
e1 = 1

s2 = 2
e2 = 5
# Extract the segment
segment1 = y[int(s1 * sr):int(e1 * sr)]
segment2 = y[int(s2 * sr):int(e2 * sr)]

fig,ax = plt.subplots(nrows=2,figsize=(15,15))

# Display the waveform
librosa.display.waveshow(segment1,ax=ax[0])
plt.title('Speech Signal Segment1')

librosa.display.waveshow(segment2,ax=ax[1])
plt.title('Speech Signal Segment2')
```

```python
# Play the segment
ipd.Audio(segment1, rate=sr)
```

```python
ipd.Audio(segment2, rate=sr)
```

# Different sampling rates

```python
fig,ax = plt.subplots(nrows=3,figsize=(15,15))
librosa.display.waveshow(y, sr=sr, ax=ax[0], label='Original')
ax[0].set_xlim(left=-0.5, right=5)
librosa.display.waveshow(y, sr=30000, ax=ax[1], label='30kHz')
ax[1].set_xlim(left=-0.5, right=5)
librosa.display.waveshow(y, sr=60000, ax=ax[2], label='60kHz')
ax[2].set_xlim(left=-0.5, right=5)
```
